### 3sum

```python
nums = [-1,0,1,2,-1,-4]
res = []
```

- sort the array

```python
nums.sort()
```

```python
for i in range(len(nums)):
    if i > 0 and nums[i] == nums[i-1]:
        continue
    
    j = i + 1
    k = len(nums) - 1
    
    while j < k:
        total = nums[i] + nums[j] + nums[k]
        if total > 0:
            k -= 1
        elif total < 0:
            j += 1
        else:
            res.append((nums[i], nums[j], nums[k]))
            while nums[j] == nums[j- 1] and j < k:
                j += 1

print(res)
```

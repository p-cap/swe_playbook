### Container with MOST water

```python
height = [1,8,6,2,5,4,8,3,7]
```

```python
left = 0
right = len(height) - 1
max_v = 0

while left < right:
    width = right - left
    h = min(height[left], height[right])
    current_area = h * width
    max_v = max(max_v, current_area)
    
    if height[left] < height[right]:
        left += 1
    else:
        right -= 1

print(max_v)
```

```python
height = [1,8,6,2,5,4,8,3,7]
left, right = 0, len(height) - 1
max_area = 0
```

```python
while left < right:
    min_height = min(height[left], height[right])
    max_area = max(max_area, (right - left) * min_height)
    if height[left] < height[right]:
        left += 1
    else:
        right -= 1
print(max_area)
```

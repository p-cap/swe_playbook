#### 128. Longest Consecutive Sequence

BREAKDOWN:
- pass the nums array into the set()
- iterate through each element
- if the element - 1 is not in num_set, we set the length to 1
- we check elements that are in num_set that are plus length away from original element
- take the longest via max between longest and length 

```python
nums = [9,1,4,7,3,-1,0,5,8,-1,6]
expected = 7
length = 0
nums_set = set(nums)   
longest = 0
```

- iterating though nums
- check if the current nums has any number before it
- check in nums if the element plus length exists within num_set

```python
for n in nums:
    if n - 1 not in nums_set:
        length = 1
    while n + length in nums_set:            
        length += 1
    longest = max(longest, length)
print(f'longest: {longest}')
print(longest == expected)
```

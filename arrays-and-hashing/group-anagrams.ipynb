#### Group Anagrams 
###### https://leetcode.com/problems/group-anagrams/description/

Approach:
- use dictionary to store anagrams with the same key
- cast the values function return into a list

```python
from collections import defaultdict

test  = ["eat","tea","tan","ate","nat","bat"]
ans = defaultdict(list)
```

- Using sorted() function to sort the characters of the string 
- the return of the sorted() is an array
- Called join function to change array to string
- append the key, even though it does not exists, to the ans defaultdict

```python
for s in test:
    key = ''.join(sorted(s))
    ans[key].append(s)
```

```python
print(ans)
```

cast the values of the ans defaultdict into a list

```python
print(list(ans.values()))
```

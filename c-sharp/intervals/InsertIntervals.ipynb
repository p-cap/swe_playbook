### 57. Insert Interval
- initialize a `list` rather than an `array`
- we'll be using a while loop therefore we will initialize `i` and `n`
    - `i` is used for iteration in the while loop
    - `n` is set as the length of the interval
- the first while loop will add all `intervals` are smaller than the `newInterval`. Because all intervals and newInterval is sorted, the second element of each `intervals` element is the larger number therefore we compare the second element to the first element of the `newInterval`
- we will merge overlapping intervals between the each of the `intervals` vs the `newInterval` elements    
    - check the Min between the newInterval first element and the `intervals` current first element
    - check the Max between the newInterval second element and the `intervals` current second element
- add the potentially modified `newInterval`
- add the remaining `intervals`
- convert the list to an array via `ToArray()` method
    - we called the `ToArray()` because the return for this problem is `int[][]` which is a jagged array

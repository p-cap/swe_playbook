### Permutation in String

#### Solution 1
- Edge case - return false if s1 is greater than s2
- initialize two arratys with a size of 26 that track character count for both s1 and s2
- BUILD THE ARRAY BASED ON THE LENGTH OF S1'S LENGTH
- iterate again based on the Window 
- the Window is computed by subtracting s1 length from s2 length
- Using the Enumerable.SequenceEqual(), check if both arrays are equal and if they are, return true
- otherwise, remove left char and add right char from s2
- return Enuemerable.SequenceEqual()

```
// set the frequency array for both s1Count and s2Count
// keep in mind, we are limited to the length of s1 
// this sets up both arrays being the same size for comparison
string s1 = "ab";
string s2 = "eidbaooo";
int[] s1Count = new int[26], s2Count = new int[26];
// this for loop builds the entire sCount
for (int i = 0; i < s1.Length; i++)
{
    // s1[0] = a - 'a'
    // s1Count[0]++
    // s1Count[0] = 1???
    s1Count[s1[i] - 'a']++;

    // s2[0] = 'e' - 'a'
    // s2[0] = 4
    // s2Count[4]++
    // s2Count[4] = 1
    s2Count[s2[i] - 'a']++;
}
// use this syntax frequently
// Console.WriteLine("=== s1Count ===");
// Console.WriteLine(string.Join(", ", s1Count));
// Console.WriteLine("=== s2Count ===");
// Console.WriteLine(string.Join(", ", s2Count));
```

```
// string s1 = "ab";
// string s2 = "eidbaooo";
// for loop is limited based on the size of the fixed Window and s2
for (int i = 0; i < s2.Length - s1.Length; i++)
{
    if (Enumerable.SequenceEqual(s1Count, s2Count))
    {
      Console.WriteLine(myresult);
      break;  
    } 
    // s2[0] - 'a' 
    // 'e' - 'a' = 4
    // s2Count[4] = 1
    // s2Count[4] = 0
    s2Count[s2[i] - 'a']--;

    // s2[0 + 2] - 'a'
    // s2[2] = b - 'a'
    // s2Count[1]
    s2Count[s2[i + s1.Length] - 'a']++;
}

Console.WriteLine("=== s1Count ===");
Console.WriteLine(string.Join(", ", s1Count));
Console.WriteLine("=== s2Count ===");
Console.WriteLine(string.Join(", ", s2Count));
```

### Minimum Window Substring

- string.IsNullOrEmpty tracks our edge case for empty strings for either s or t
- target is initialized as an array with size of 128
- With 128 in size, I think we are tracking most ASCII characters 
- we set the target array with the characters from t string
- NOTE: to get the decimal value of an ASCII character, we cast it with an int
- Iterate through s string right pointer
- if the rChar, character of the right pointer,  target array and if it rChar number in the window array is less than the number in the target array, we increment the currentMatchCount
- add rChar to the window array
- check if the currentMatchCount is equal to the requiredCount
- requiredCount is essentially set by the length of t copunt
- if the currentMatchCount is equal to requiredCount it enters the while loop
- If the position of the right and left pointer plus one is less than `minLen`, we set those positions to minLen
- we set the left pointer to the startIndex
- outside the if statement, we set the lChar
- this time, we check for the lChar in target is greater than 0 and if lChar in window is less than the lChar in target
- if so, we decrease the currentMatchCount
- outside the lChar checker if statement, we increment with left pointer
- the main while loop will always increment right pointer

```polyglot-notebook
int[] window = new int[128];
int left = 0, right = 0;
int requiredCount = t.Length; // Total chars to match
int currentMatchCount = 0;    // Chars we have successfully matched
int minLen = int.MaxValue;
int startIndex = 0;
```

```polyglot-notebook
string s = "ADOBECODEBANC", t = "ABC";
string expected = "BANC";

while (right < s.Length)
{
    // what is rChar is used?
    // rChar starts at the beginning of the string
    // rChar is "A"
    // right now is 1
    // the next rChar is "D"
    char rChar = s[right];

    // target[rChar] is currently not greater than 0
    // window[rChar] is not less than target[rChar]
    // it looks like rChar in window tracks differently than target
    // we are not satisfying the first condition with the first iteration
    // be aware that target was previously set
    // target = []
    // foreach (char c in t) target[c]++;
    // target = []
    if (target[rChar] > 0 && window[rChar] < target[rChar]) {
        currentMatchCount++;
    }

    // we are increasing the currnt rChar in the window array
    // int[] window = new int[128]
    // uppercase is smaller in value
    // lowercase is higher in value
    // window['A'] is at 65
    // window[65] = 1
    // window now has 
    window[rChar]++;

    // currentMatchCount = 0
    // requiredCount is the length of t
    // why do we use equality between currentMatchCount and requiredCount????
    // currentMatchCount is currently NOT equal to requiredCount
    // requiredCount is the length of t
    while (currentMatchCount == requiredCount)
    {
        if (right - left + 1 < minLen)
        {
            minLen = right - left + 1;
            startIndex = left;
        }

        char lChar = s[left];
        window[lChar]--;

        if (target[lChar] > 0 && window[lChar] < target[lChar])
        {
            currentMatchCount--;
        }
        left++;
    }

    // increase right
    // right will increase regardless
    // right = 1
    // right is part of the while loop expression
    right++;
}

string result  = minLen == int.MaxValue ? "" : s.Substring(startIndex, minLen);
Console.WriteLine(result);
```

### REDO

- given the input s and t, find the minimum substring in s based on t

```polyglot-notebook
string s = "ADOBECODEBANC", t = "ABC";
string expected = "BANC";
int left = 0, right = 0;
int currentMatchCount = 0;
int requiredCount = 0;

// there are two arrays used
int[] window = new int[128];

// setting target to account for letters in t 
int[] target = new int[128];
foreach (char ch in t) target[ch]++;

int minLen = int.MaxValue;
int startIndex = 0;

if (string.IsNullOrEmpty(s) || string.IsNullOrEmpty(t)) Console.WriteLine(0);

while (right < s.Length)
{
    char rChar = s[right];

    // target[rChar] > 0 -> checks if rChar exists in target
    // window[rChar] < taregt[rChar] -> makes sure that the number of rChar characters are not more than what's in target
    if (target[rChar] > 0 && window[rChar] < target[rChar])
    {
        // currentMatchCount accounts for characters on both s and t
        currentMatchCount++;
    }

    // window accounts for all characters
    window[rChar]++;

    // currentMatchCount will reach requiredCount only if it has seen all characters
    while (currentMatchCount == requiredCount)
    {
        if (right - left + 1 < minLen)
        {
            minLen = right - left + 1;
            startIndex = left;
        }

        // account for the left pointer
        // because the currentMatch is equal to requiredCount
        char lChar = s[left];
        window[lChar]--;

        if (target[lChar] > 0 && window[lChar] < target[lChar])
        {
            currentMatchCount--;
        }
        left++;
    }

    right++;
}

string result  = minLen == int.MaxValue ? "" : s.Substring(startIndex, minLen);
Console.WriteLine(result);
```

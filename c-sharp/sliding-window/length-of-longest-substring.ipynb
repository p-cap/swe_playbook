### Length of Longest Substring 

- initialize an array with a size of 128 which accounts for all ASCII characters
- fill all the elements with -1
- sliding window is the algorithm
- left and right are initialized at zero
- iterate through the input string
- if the current string is greater than or equal to the current left pointer, we update the left pointer 
    - Why? If the charater is a duplicate, the character index will have a value greater than left 
- otherwise, the character index is updated with the right value
- we calculate the max between the current max and the distance between right and left plus one which is the window

```

string s = "abcabcbb";
int n = s.Length;
int maxLength = 0;

int[] lastIndex = new int[128]; 
Array.Fill(lastIndex, -1);

int left = 0;

for (int right = 0; right < n; right++)
{
    // right = 0
    char currentChar = s[right];
    // currentChar = 'a'
    // in C#, when a char is 16-bit numeric value under the hood
    // When we use it as an index, the computer performs an implicit conversion
    // from a character to its corresponding integer (UTF-16/ASCII code)
    // in this case, lastIndex[currentChar], currentChar is implicitly converted
    if (lastIndex[currentChar] >= left)
    {
        left = lastIndex[currentChar] + 1;
    }

    lastIndex[currentChar] = right;
    maxLength = Math.Max(maxLength, right - left + 1);
}
Console.WriteLine($"maxLength: {maxLength}");
```

### Best time to buy and sell stock

How does the sliding window move?
- left pointer is initialized to 0 and the right pointer is initilized to 1.
- if the left pointer is greater than the right pointer, the left is now eqaul to the right
- rigth is alway moving because it looks for the next number to compare to left

How is  `max_profit` calculated?
- if the right pointer is greater than the left pointer, we calculate the max_profit

```polyglot-notebook
public class Solution {
    public int MaxProfit(int[] prices) {

        int max_profit = 0;

        // initalize the smallest window to start
        int left = 0, right = 1;
        int len = prices.Count();
        
        while (right < len) {
            // if right is greater than left, then we calculate max_profit
            if (prices[right] > prices[left]) {
                max_profit = int.Max(max_profit, prices[right] - prices[left]);
            } else {
                // we resize the window if left is greater than right
                left = right;
            }
            // right pointer moces regardless
            right += 1;
        }
        return max_profit;
    }
}
```

```polyglot-notebook
var mySolution = new Solution();
int[] prices = [7,1,5,3,6,4];
int expected = 5;
Console.WriteLine(mySolution.MaxProfit(prices) == expected);
```

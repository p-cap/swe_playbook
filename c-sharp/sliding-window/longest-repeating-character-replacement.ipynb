### Longest Repeating Character Replacement

- declare `counts` array to account for all uppercase letters
- account for the `maxFreq` and the `maxLength`
- iterate with the `right` pointer
- calculate the currentIndex by subtracting the current char with 'A'
- increase the counts of the index
- check if the window minus maxFeq is greater than left
- if so, we decrement the left pointer character
- and, move the left pointer
- take the current `maxLength` and the size of the current window
- return the `maxLength`

```polyglot-notebook
string s = "ABAB";
int k = 2;
int expected = 4;
```

```polyglot-notebook
// frequency map A- Z
// how is a frequency map tracking substrings
int[] counts = new int[26];
int left = 0;
// the count of most frequent characters
int maxFreq = 0;
int maxLength = 0;
```

```polyglot-notebook
// string s = "ABAB";
// int k = 2;

// s.Length = 4
for (int right = 0; right < s.Length; right++) {
    // 1. expand the window: Add new character

    // s[0] - 'A' = 'A' - 'A' = 0
    // s[1] - 'A' = 'B' - 'A' = 1
    // s[2] - 'A' = 'A' - 'A' = 0
    // s[3] - 'A' = 'B' - 'A' = 1
    int currentIndex = s[right] - 'A';
    // currentIndex = 0
    // currentIndex = 1
    // currentIndex = 0
    // currentIndex = 1

    // counts[0]++
    // counts[1]++
    // counts[0]++ = 1 + 1 = 2
    // counts[1]++ = 1 + 1 = 2
    counts[currentIndex]++;
    // counts[0] = 1
    // counts[1] = 1
    // counts[0] = 2
    // counts[1] = 2

    // 2. update the max frequency found in the current window
    // Math.Max(0,1)
    // Math.Max(1,1)
    // Math.Max(1,2)
    // Math.Max(2,2)
    maxFreq = Math.Max(maxFreq, counts[currentIndex]);
    // maxFreq = 1
    // maxFreq = 1
    // maxFreq = 2
    // maxFreq = 2

    // 3. Check if the window is valid????
    // (0 - 0 + 1) - 1 > 2 
    // 0 > 2
    // (1 - 0 + 1) - 1 > 2
    // 1 > 2 
    // (2 - 0 + 1) - 2 > 2
    // 1 > 2
    // (3 - 0 + 1) - 2 > 2
    while ((right - left + 1) - maxFreq > k) {
        counts[s[left] - 'A']--;
        left++;
    }    

    // Math.Max(0, 1)
    // Math.Max(1, 1 - 0 + 1) = 2
    // Math.Max(2, 2 - 0 + 1) = 3
    // Math.Max(3, 3 - 0 + 1) = 4
    maxLength = Math.Max(maxLength, right - left + 1);
    // maxLength = 1
    // maxLength = 2
    // maxLength = 3
    // maxLength = 4
}

Console.WriteLine(maxLength == expected);
```

156. Find Minimum Rotated Sorted Array

- Binary Search 
- left and right pointer
- if mid is greater than right, we move the left pointer
- otherwise, move the left pointer
- THE CONDITION OF THE WHILE LOOP IS `Left < right` 
- after the loop, left == right pointing to the minimum

```
int[] nums = [4,5,6,7,0,1,2];
```

```
int left = 0;
int right = nums.Length - 1;
int expected = 0;
```

```
while (left < right)
{
    int mid = left + (right - left) / 2;
    if (nums[mid] > nums[right])
    {
        left = mid + 1;
    } else
    {
        right = mid;
    }
}

Console.WriteLine(nums[left] == expected);
```

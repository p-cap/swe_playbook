### 704 Binary Search

 - List declaration

```
// lists
// there are other conventions of declaring a list but I like this one
// the convention below is NOT fixed
List<int> nums = [];
for (int i = 0; i < 10; i++)
{
    nums.Add(i);
}

foreach (int i in nums)
{
    Console.Write(i);
}
```

```
// while loop in c sharp
int left = 0;
int right = nums.Count - 1;
Console.WriteLine(right);

while (left < right)
{
    Console.WriteLine("We are good");
    left += 1;
}
```

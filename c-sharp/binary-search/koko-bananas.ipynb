### Koko Eating Bananas

- the algorithm is binary search
- we'll have two pointers, `low` and `high`
- low is assigned as 1, why? because 1 is lowest possible value which is the lowest number of bananas that Koko can eat
- our `high` pointer is the maximum number in the list
- the expression is `low <= high`. We are result is returning a single number, we can use the `<=` sign
- we assign the result to the same as the high pointer? Why? becaues the max number is the maximum number of bananas that Koko can eat amonsgt the pile of bananas
- k is essentially our mid for the problem
- we need to account for totalTime. Total time is tracked so that we can use it to check it with h. 
- start iterating each pile
- calculate the total time is the current pile plus k plus 1 divided by k?????
- if the total time is less than h
    - set k as the result
    - high is moved 
- else we move low???

```
int[] piles = [3,6,7,11];
int h = 8;
int expected = 4;
```

```
using System.Linq;

int low = 1;
int high = piles.Max();
int result = high;

while (low <= high)
{
    int k = (low + high + 1) / 2;

    int totalTime = 0;

    foreach (int p in piles)
    {
        totalTime += (p + k - 1) / k;
    }

    if (totalTime <= h)
    {
        result = k;
        high = k - 1;
    }

    else
    {
        low = k + 1;
    }
}

Console.WriteLine(result == expected);
```

Koko Eating Bananas

```
int[] piles = [3,6,7,11];
int h = 8;
int expected = 4;
```

```
using System.Linq;

int low = 1;
int high = piles.Max();
int result = high;

Console.WriteLine(high);

while (low <= high)
{
    int k = (low + high + 1) / 2;

    int totalTime = 0;

    foreach (int p in piles)
    {
        totalTime += (p + k - 1) / k;
    }

    if (totalTime <= h)
    {
        result = k;
        high = k - 1;
    }

    else
    {
        low = k + 1;
    }
}

Console.WriteLine(result == expected);
```

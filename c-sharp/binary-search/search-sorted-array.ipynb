### Search in rotated array

- binary search
- declare `high` and `low` pointer
- check if mid is equal to the target
- check if low is less than mid
    - check if target is greater low and is less than mid
    - if so, move the `high` pointer
    - otherwise, move the `low`
- check if the target is greater than mid and less than high
- if so, move the `low` pointer 
- else move the `high` pointer


```
int[] nums = [4,5,6,7,0,1,2];
int target = 4;
int expected = 4;
```

```
int result = -1;
int low = 0, high = nums.Length - 1;

while (low <= high)
{
    int mid = low + (high - low) / 2;
    if (nums[mid] == target)
    {
        result = mid;
        break;
    }
    // equal sign do matter
    if (nums[low] <= nums[mid])
    {

        // equal sign do matter
        if (target >= nums[low] && target < nums[mid])
        {
            high = mid - 1;
        }
        else
        {
            low = mid + 1;
        }
    }
    else
    {
        // equal sign do matter
        if (target > nums[mid] && target <= nums[high])
        {
            low = mid + 1;
        }
        else
        {
            high = mid - 1;
        }
    }
}

Console.WriteLine(result);
```

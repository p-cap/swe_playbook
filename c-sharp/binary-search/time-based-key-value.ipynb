### Time-based Key-Value Store

- declare the TimeMap class
    - dictionary with a string as the ket and the list/tuple of string and int as the value
    - in the constructor, initialize a new record store
- inside the set method
    - if there is no key in the record, add create a key with the new list/tuple instance
    - always add a value to the key being the key has a list of values
- inside the get method
    - check if the key exists
    - if there is only one element in the list of tuples
        - if the record timestamp is less than the timestamp being asked, return that timestamp
        - else return the result which is empty
    - if the record count is 2
        - check the second record if it is less than the timestamp being passed on, return thet value of that record
        - else if the first record is less than than the timestamp being passed on, return that value of that record
        - else return reult
    - perform a binary search
        - if the mid value is greater than the timestamp, move the right pointer
        - if the mid value is less than the timestamp, move the left pointer
        - return the mid value
    - if the right timestamp is less than the timestamp, return the right value
    - return the result

```
public class TimeMap
{
    private Dictionary<string,List<(string value, int timestamp)>> records;

    // make sure I am declaring a constructor
    // the constructor has a ()
    public TimeMap() {
        records = new();
    }

    public void Set(string key, string value, int timestamp)
    {
        if (!records.ContainsKey(key))
        {
            // removed the new keyword for initialization
            // instead of new List<(string value, int timestamp)>()
            records.Add(key, new List<(string value, int timestamp)>());
        }
        records[key].Add((value, timestamp));
    }

    public string Get(string key, int timestamp)
    {
        // the incoming timestamp should be greater than the timestamp 
        var result = "";

        // this checkts if the key exists, and if it is, return the record variable
        if (records.TryGetValue(key, out var record))
        {
            if (record.Count == 1)
            {
                if (record[0].timestamp <= timestamp) { return record[0].value; }
                else return result;
            }

            if (record.Count == 2)
            {
                if (record[1].timestamp <= timestamp) { return record[1].value; }
                else if (record[0].timestamp <= timestamp) { return record[0].value; }
                else return result;
            }

            int left = 0;
            int right = record.Count - 1;

            while(left <= right)
            {
                int middle = left + (right - left) / 2;

                if (record[middle].timestamp > timestamp) { right = middle - 1;}
                else if (record[middle].timestamp < timestamp) {left = middle + 1;}
                else {return record[middle].value;}
            }

            if (record[right].timestamp < timestamp) { return record[right].value; }
        }

        return result;
    }
}
```

```
var timeMap = new TimeMap();
timeMap.Set("foo","bar",1);
Console.WriteLine(timeMap.Get("foo",1));
```

#### 74. Search a 2D Matrix

- declare test case

```
int[][] matrix = [[1],[3]];
int target = 3;
```

- we'll use binary search
- we'll check if the target falls to what row

```
using System.Numerics;
using System.Runtime;

int top = 0;
int bottom = matrix.Count() - 1;
int start = 0;
// let's make sure that the last element has a -1
int end = matrix[0].Count() - 1;
int selected_row = 0;

// the equality handles a 1 x 1 matrix
while (top <= bottom)
{
    int mid = (top + bottom + 1) / 2;
    if (target >= matrix[mid][start] & target <= matrix[mid][end])
    {
        Console.WriteLine("Found the row");
        selected_row = mid;
        break;
    }

    else if (target < matrix[mid][start])
    {
        bottom = mid - 1;
    }

    else
    {
        top = mid + 1;
    }
}

Console.WriteLine($"selected_row: {selected_row}");
```

- THE CULPRIT: I incorrectly assigned the last index of an array by setting `int end = matrix.Count()`
- I needed to initialize it as `int end = matrix.Count() - 1`
- IMPORTANT: `Make sure to put a break on your while loop to avoid infinite loop because this instance of notebbok tends to hang`

```
int start = 0;
int end = matrix[0].Count() - 1;
bool found = false;

// handles 1 element rows
// always have a break on while loop
while (start <= end)
{
    int mid = (start + end + 1) / 2;
    if (matrix[selected_row][mid] == target)
    {
        found = true;
        Console.WriteLine($"matrix[selected_row][mid]: {matrix[selected_row][mid]}");
        break;
    } 
    else if (target > matrix[selected_row][mid]) {
        start = mid + 1;
    }
    else
    {
        end = mid - 1;
    }
}

if (!found)
{
    Console.WriteLine("No match of the target was found");    
}
```

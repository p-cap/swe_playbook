### Contain Duplicates
- pass the `nums` into `HashSet` instantiation
```csharp
HashSet<int> trackDupes = new(nums);
```
- if the length of nums not equal to hashset, nums contain duplicates
```csharp
return nums.Length != trackDupes.Count()
```

````python
public class Solution {
    public bool ContainsDuplicate(int[] nums) {
        HashSet<int> trackDupes = new(nums);
        return nums.Length != trackDupes.Count();
    }
}
````

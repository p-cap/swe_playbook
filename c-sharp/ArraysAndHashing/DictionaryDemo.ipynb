### Dictionary<TKey,TValue> Class

## Dictionary

- Initialize dictionary
- call the `Add` function to add elements to dictionary
- use `foreach` to iterate through the dictionary

```polyglot-notebook
Dictionary<char, int> test = [];
test.Add('c',1);

foreach (var (ch, val) in test) {
    Console.WriteLine($"{ch} : {val}");
}
```

- `TryGetValue` method
- checks if the key is available
- if key exixts, it will return value otherwise 0

```polyglot-notebook
test.TryGetValue('b', out int firstCount);
Console.WriteLine(firstCount);
test.TryGetValue('c', out int secondCount);
Console.WriteLine(secondCount);
```

- printing dictionary values

```polyglot-notebook
foreach (int num in test.Values) {
    Console.WriteLine(num);
}
```

- printing dictionary keys

```polyglot-notebook
foreach (char ch in test.Keys) {
    Console.WriteLine(ch);
}
```

- check for keys in dictionary

```polyglot-notebook
if (test.ContainsKey('c')) {
    Console.WriteLine("The dictionary contains 'c' ");
}
```

- check values in dictionary

```polyglot-notebook
if (test.ContainsValue(1)) {
    Console.WriteLine("The dictionary contains 1 ");
}
```

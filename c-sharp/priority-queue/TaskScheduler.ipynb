### 621. Task Scheduler

```polyglot-notebook
char[] tasks = ['A','A','A','B','B','B'];
int n = 2;

if (n == 0) Console.WriteLine(tasks.Length);

// 1. Count frequencies
int[] counts = new int[26];
foreach (char t in tasks) {
    counts[t - 'A']++;
}

// Max-Heap to store frequencies
var maxHeap = new PriorityQueue<int,int>();
// the negative priority allows us set the max heap
foreach (int count in counts) {
    if (count > 0) maxHeap.Enqueue(count, -count);
}

int time = 0;
Queue<(int count, int availableAt)> cooldownQueue = new();

while (maxHeap.Count > 0 || cooldownQueue.Count > 0) {
    time++;

    // process task from Heap
    if (maxHeap.Count > 0) {
        int remaining = maxHeap.Dequeue() - 1;
        if (remaining > 0) {
            cooldownQueue.Enqueue((remaining, time + n));
        }
    }

    if (cooldownQueue.Count > 0 && cooldownQueue.Peek().availableAt == time) {
        int readyCount = cooldownQueue.Dequeue().count;
        maxHeap.Enqueue(readyCount, -readyCount);
    }
}

Console.WriteLine(time);
```

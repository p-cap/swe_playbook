### 703. Kth Largest Element in a Stream
#### ALGO
- initialize `PriorityQueue`
- in the constructor:
    - EDGE: if nums is null, return
    - iterate through input array and call the `Add` method
- Add method
    - if the PriorityQueue is less than k, we add value to the queue:
    - `_minHeap.Enqueue(val,val);`??????
    - if the value being added is greater than `_minHeap.Peek()`, we call `_minHeap.EnqueueDequeue(val,val)`
    - return `_minHeap.Peek()`


```polyglot-notebook
public class KthLargest
{
    private readonly PriorityQueue<int,int> _minHeap;
    private readonly int _k;
    
    public KthLargest(int k, int[] nums)
    {
        _k = k;
        _minHeap = PriorityQueue<int,int>();

        if (nums == null) return;

        foreach (int num in nums) 
        {
            Add(num);
        } 
    }

    public int Add(int val)
    {
        if (_minHeap < _k)
        {
            _minHeap.Enqueue(val,val);
        } 
        else if (val > _minHeap.Peek())
        {
            _minHeap.EnqueueDequeue(val,val);
        }

        return _minHeap.Peek();
    }
}
```

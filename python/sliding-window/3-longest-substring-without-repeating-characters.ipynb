### 3. Longest Substring Without Repeating Characters
- same direction two pointer
- track seen characters hashset
- iterate through string with for loop using right pointer
- if `right` pointer is in set, we remove left pointer element in a WHILE LOOP
- add all seen right elements
- calculate the max window from the left and right pointer

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        """
        - same direction two pointer
        - iterate through s using for loop
        - if left pointer is the same as the right pointer
            - remove the left character from the hashset
            - move the left pointer
        - right pointer is iterated through for loop
        - add the right pointer onto the hashset
        - take the longest substring based on the current max and the distance between right and left pointer 
        """

        left = 0
        n = len(s)
        seen = set()
        substring = 0
        for right in range(n):
            print(f'left:  {left}')
            print(f'right: {right}')
            # why is this significantly important?
            while s[right] in seen:
                seen.remove(s[left])
                left += 1                
            seen.add(s[right])
            substring = max(substring, right - left + 1)
        return substring
    
s = "pwwkew"
my_solution = Solution()
output = my_solution.lengthOfLongestSubstring(s)
print(output)
```

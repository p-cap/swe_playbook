### Permutation in string

```python
s1 = "ab", s2 = "eidbaooo"
n1, n2 = len(s1), len(s2)
result = True
if n1 > n2: result = False

# Initialize frequency list
s1_count = [0] * 26
window_count = [0] * 26

# fill the target and first window
for i in range(n1):
    s1_count[ord(s1[i]) - ord('a')] += 1
    window_count[ord(s2[i]) - ord('a')] += 1
    

# check the current s1_count and window_count
if s1_count == window_count:
    print(True)
    
for i in range(n1, n2):
    # add the new char to window
    window_count[ord(s2[i]) - ord('a')] += 1
    
    # remove char from the window
    window_count[ord(s2[i - n1]) - ord('a')] -= 1
    
    # check the current window
    if s1_count == window_count:
        print(True)
        break
    
```

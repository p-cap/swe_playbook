### Permutation in string

```python
s1 = "ab", s2 = "eidbaooo"
n1, n2 = len(s1), len(s2)
result = True
if n1 > n2: result = False

# Initialize frequency list
s1_count = [0] * 26
window_count = [0] * 26

# fill the target and first window
for i in range(n1):
    s1_count[ord(s1[i]) - ord('a')] += 1
    window_count[ord(s2[i]) - ord('a')] += 1
    

# check the current s1_count and window_count
if s1_count == window_count:
    print(True)
    
for i in range(n1, n2):
    # add the new char to window
    window_count[ord(s2[i]) - ord('a')] += 1
    
    # remove char from the window
    window_count[ord(s2[i - n1]) - ord('a')] -= 1
    
    # check the current window
    if s1_count == window_count:
        print(True)
        break
    
```

### Redo

```python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:        
        """
        - EDGE CASE: if s1 is greater than s2, return False
        - Build the frequency map for:
            - s1 count 
            - window
        - check if the current window and s1_count are equal and if so, return True
        - iterate through s2 starting with length of s1
        - remove the char from the current window freq map
        - add the current char to the same current window freq map
        - check if the current window and s1 count are equal and if so, return True
        - return False as the default for this function
        """
        n1, n2 = len(s1), len(s2)

        if n1 > n2: return False

        s1_count = [0] * 26
        window = [0] * 26
        
        for i in range(n1):
            s1_count[ord(s1[i]) - ord('a')] += 1
            window[ord(s2[i]) - ord('a')] += 1
        if s1_count == window:
            return True

        for i in range(n1,n2):
            window[ord(s2[i-n1]) - ord('a')] -= 1
            window[ord(s2[i]) - ord('a')] += 1

            if s1_count == window:
                return True
        
        return False
    
s1 = "ab"
s2 = "eidboaoo"

my_solution = Solution()
result = my_solution.checkInclusion(s1, s2)
print(result)
```

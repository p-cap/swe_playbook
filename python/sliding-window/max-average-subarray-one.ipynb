### 643. Maximum Average Subarray I
- check for the initial sum of the first window 
- iterate through the for loop. 
- IMPORTANT: We need to start from k as the starting point for the for loop
- remove the first element by subtracting from the sum
- add current element to the sum

```python
nums = [1,12,-5,-6,50,3]
k = 4
max_sum = current_sum = sum(nums[:k])
n = len(nums)

for i in range(k, n):
    print()
    current_sum = current_sum - nums[i-k] + nums[i]
    max_sum = max(max_sum, current_sum)
print(max_sum / k)
```

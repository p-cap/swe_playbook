Number of Connected Components in an Undirected Graph
- create an adj graph -> both edges become key and value to each other
- tracked visited via set
- initialize number of compoenents
- declare a recursive function -> the purpose is to check neighbors of the current node and tracks them as visited
- the function is called recursively so that it can check other neighbor of neighbors
- if not visited, add it to the visited set and call the recursive function
- in the main function, iterate through each node
- if not visited, increment component variable
- call dfs

```python
n = 5
edges = [[0,1],[1,2],[3,4]]
adj = {i: [] for i in range(n)}

for u,v in edges:
    adj[u].append(v)
    adj[v].append(u)

print(adj)

visited = set()
components = 0

# checks the neighbors sof the node and adds it to the visited
def dfs(node):
    for neighbor in adj[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            dfs(neighbor)

# iterate through each node based on n 
for i in range(n):

    # if not visited, we increment component by 1
    if i not in visited:
        components += 1
        visited.add(i)
        dfs(i)

print(components)
```

NOTE:
- the number components is determined by:
    - if the node has not been visited
    - if it's neighbors has been visited

### Cloned Graph
- ALGO: DFS
- initialize hash map for tracking
- DFS function will:
    - BASE CASE: check if the node is None 
    - check if the node exists within the hash map and if so, return the node within the hash map
    - create a copy 
    - set that copy to the hash map
    - recursively look through the neighbors
    - append the necessary neighbors to the current node
    - return the copy

```python
old_to_new = {}

def dfs(node):
    if not node: # BASE CASE: check if node is None
        return None
    
    if node in old_to_new: # look node in old_to_new
        return node
    
    copy = Node(node.val) # generate a copy of the Node
    old_to_new[node] = copy # set the copy to the old_to_new for tracking

    for neighbor in node.neighbors: # check the neighbors of the current node
        copy.neighbors.append(dfs(neighbor)) # append the neighbors by calling dfs
    
    return copy # return the copy

dfs(node)
```

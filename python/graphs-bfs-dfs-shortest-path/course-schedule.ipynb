```python
numCourses = 2
prerequisites = [[1,0]]

# 1. Build an adjacency list
adj = {i: [] for i in range(numCourses)}
for crs, pre in prerequisites:
    adj[crs].append(pre)
print(adj)

# # TESTING: iterating through list of lists
# test = [[3,3],[4,4]]
# for first,second in test:
#     print(f'first: {first}')
#     print(f'second: {second}')

# visitState: 0 = unvisited, 1 = visiting, 2 = visited
visit = [0] * numCourses

def dfs(crs):
    if visit[crs] == 1: # found a cycle?
        return False
    if visit[crs] == 2:
        return True
    
    visit[crs] = 1 # Mark as currently visiting

    # looping through the adjacency list
    for pre in adj[crs]:
        if not dfs(pre):
            return False

    visit[crs] = 2 # Mark as fully safe???
    return True

for i in range(numCourses):
    if not dfs(i):
        print(False)
        break
```

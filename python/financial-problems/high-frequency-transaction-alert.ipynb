High-Frequency Transaction Alert

Scenario: Given a list of transaction amounts and a window size $k$, find the maximum average transaction value within any contiguous window of size $k$. If the average exceeds a threshold, we flag it.

```python
from collections import defaultdict

amounts = [100, 200, 300, 400, 100]
k = 3 
threshold = 250
current_window = amounts[:k]
current_sum = sum(current_window) 
max_average = current_average = current_sum / k
n = len(amounts)
current_average_to_window = defaultdict(list)

for i in range(k, n):
    current_sum = (current_sum - amounts[i-k] + amounts[i]) 
    max_average = max(max_average, current_sum / k)

if max_average > threshold:
    print(max_average)
```

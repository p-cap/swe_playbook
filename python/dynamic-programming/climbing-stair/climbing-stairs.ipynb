### Climbing Stairs
#### Pure recursion
- create a base case that returns the values
- create the recursive relation
- return the recursive relation 
- call the recursive function
- it is a Time exceeded solution according to leetcode

```python
n = 7
expected = 21

def recurse(n):
    if n == 1: return 1
    elif n == 2: return 2
    return recurse(n - 1) + recurse(n-2)

recurse(n) == expected
```

#### Recursion with memoization
- create the recursive function
- initialize a defaultdict
- return the necessary base cases
- add the sum between the recursive calls and add it to the map
- return the defaultdict value of the key
- this is still a time exceeded solution according to leetcode

```python
from collections import defaultdict
my_map  = defaultdict(int)

n == 7
expected == 21

def recurse(n):
    if n == 1 or n == 2: my_map[n] = n
    else: my_map[n] = recurse(n-1) + recurse(n-2)
    return my_map[n]

print(recurse(7))
```

### Pure Memoization
Initialize a map to store the values on every call. Iterate through the loop based on the stair number. Save each key based on the previous values. 

```python
n = 7
expected = 21
my_map = defaultdict(int)

for i in range(1, n + 1):
    if i == 1 or i == 2:
        my_map[i] = i
    else: 
        my_map[i] = my_map[i-1] + my_map[i-2]

print(my_map)
```

### Pure iteration
I will initialize 3 variables. 2 of the 3 variables will hold the first and second number during the iteration of the loop. The third variable will store the result from the other two variables. There two edge cases we need to handle, when n = 1 and n = 2. When both are passed as inputs, we return the same value. Within the for loop, we store the first two numbers. Once the first two numbers are stored, we add the first two numbers to the resulting variable. The first number stores the second number and the second number stores the resulting variable. We return the resulting variable as the final result

```python
first, second, stairs = 0,0,0
#  we are not calling a function there I will not handle the edge case here
# if n == 1:
#     return 1
# elif n == 2:
#     return 2    
n = 7
expected = 21
for i in range(1, n + 1):
    if i == 1:
        first = 1
    elif i == 2:
        second = 2
    else:
        stairs = first + second
        first = second
        second = stairs
print(stairs == expected)
```

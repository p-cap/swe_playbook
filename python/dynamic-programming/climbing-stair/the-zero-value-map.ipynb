## The zero value map 
The code I was running did not populate the correct key and value. The keys were populated with no values. The base case, line 5, was never satisfied because I did not call the function again. The else statement then calculated the current value of n by retrieving the values of n however there were no values to retrieve. In the end, the keys n, n - 1 and n - 2 were populated in the map however it did not have any values

```python
from collections import defaultdict
my_map = defaultdict(int)
def recurse(n):
    if n == 1 or n == 2: my_map[n] = n
    # {6: 0, 5: 0, 7: 0})
    else: my_map[n] = my_map[n-1] + my_map[n-2]
    return my_map[n]
print(recurse(7))
```

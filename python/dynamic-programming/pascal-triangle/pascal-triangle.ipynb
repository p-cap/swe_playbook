### Pascal's triangle

#### Recursive fashion with a global variable
I initialize the triangle to store the results of the problem. Then, I create the recursive function that take the number of rows as a parameter. The base case is if the number of rows reaches 1, then it will return [[1]] which is considered the top of the triangle. Inside the base case, we are only calling return just to terminate the function. Then, we start calling the recursive function and deducting the numRows parameter by 1. We grab the previous row from the triangle and initial the new row containing 1. We iterate through the `prev row - 1` and calculating the middle elements for the new row. IMPORTANT: the new row element is calculate from the two prev row elements. 

## IMPORTANT
Use the sample of print statements below to make sure we use print statements within the problem in question vs isolating it so we can learn how to debug our code during online assessments

```python
numRows = 5
expected = [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
triangle = []

# the generate function will ultimately return [[1]] once it hits the base case
def generate(numRows):
    global triangle
    if numRows == 1: 
        triangle = [[1]]
        return
    generate(numRows - 1)
    prev_row = triangle[-1]
    new_row = [1]
    
    # calculate the middle elements
    print(f"prev_row: {prev_row}")
    print(f"len(prev_row) - 1: {len(prev_row) - 1}")
    for j in range(len(prev_row) - 1):
        print(f"j:{j}")
        print(f"prev_row[j]:{prev_row[j]}")
        print(f"prev_row[j + 1]:{prev_row[j + 1]}")
        new_row.append(prev_row[j] + prev_row[j+1])
    print("==================")
    new_row.append(1)
    triangle.append(new_row)
generate(numRows)
print(triangle)
```

#### Recursion without the global variable    

```python
numRows = 5
expected = [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

def generate(numRows):
    if numRows == 1: return [[1]]
    # triangle is local to the function
    triangle = generate(numRows - 1)
    prev_row = triangle[-1]
    new_row = [1]
    
    # the -1 prevents the prev_row + 1 to go out of range
    # it also allows us to append the last 1 when building the new row
    for j in range(len(prev_row) - 1):
        new_row.append(prev_row[j] + prev_row[j+1])
        
    new_row.append(1)
    triangle.append(new_row)
    return triangle


result = generate(5)
print(result)
print(result == expected)
```

### Dynamic programming 
The generate function is created but only called once. We use a for loop with the numRows as the ending index. We build the rows inside the for loop. Another for loop is built but this time, we start at 1 and we end at the current index of the outer loop. We calculate the elements based on dp and set it to row's current state. The calculation is `dp[i - 1][j - 1] + dp[i - 1][j]` which is set to the current element within the row variable. We go out of the inner for loop and then we append the row into the dp

```python
numRows = 5
expected = [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
dp = []

# the global varible is recognized because I think it is called once?
def generate(numRows):
    for i in range(numRows):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = dp[i - 1][j - 1] + dp[i - 1][j]
        dp.append(row)
generate(5)
print(dp)

# result = generate(5)
# print(result)
```

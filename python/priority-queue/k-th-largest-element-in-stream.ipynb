### 703. Kth Largest Element in a Stream
- upon iniitialization of class:
    - heapify the list
    - using while loop with the expression that checks the length of input list and k
- the add method:
    - heapifies the list
    - push element onto the heap
    - checks if the length of input list is greater than nums then pop from the min heap
    - return the first element in the heap

```python
import heapq

class KthLargest(object):

    def __init__(self, k, nums):
        """
        :type k: int
        :type nums: List[int]
        """
        self.nums = nums
        self.k = k
        heapq.heapify(self.nums)
        while len(self.nums) > k:
            heapq.heappop(self.nums)
    
    def add(self, val):
        """
        :type val: int
        :rtype: int
        """
        heapq.heappush(self.nums,val)    
        if len(self.nums) > self.k:
            heapq.heappop(self.nums)

        return self.nums[0]
        
# Your KthLargest object will be instantiated and called as such:
# obj = KthLargest(k, nums)
# param_1 = obj.add(val)
```

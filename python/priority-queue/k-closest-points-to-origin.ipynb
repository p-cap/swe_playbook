973. K Closest Points to Origin

- iterate through the points
- calculated the distance of each point from 0,0
- initialized max_heap by adding the distance as negative numbers
- heapify the max_heap
- perform heappop to match the kth element
- append results wtth only the points

```python
import heapq

class Solution:
    def kClosest(self, points: List[List[int]], K: int) -> List[List[int]]:
        """
        - ALGO: max_heap
        - iterate through the points
        - calculated the distance of each point from 0,0
        - initialized max_heap by adding the distance as negative numbers
        - heapify the max_heap
        - perform heappop to match the kth element
        - append results wtth only the points
        """
        max_heap = []
        result = []
        for pt in points:
            distance = sqrt(pt[0]*pt[0] + pt[1]*pt[1])
            max_heap.append((-distance,pt[0],pt[1]))
        heapq.heapify(max_heap)
        while len(max_heap) > K:
            heapq.heappop(max_heap)
        for i in range(K):
            result.append([max_heap[i][1],max_heap[i][2]])
        return result
```

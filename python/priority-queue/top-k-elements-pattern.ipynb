### 1. Top K Elements Pattern

Goal: Find the kth largest, smallest or most frequent elements in a dataset of size
Algorithm:
- import the `Counter` module to create a frequency map
- import `heapq` to utilize heap data structure
- consider `k` which is the kth element
- iterate through the Counter instance by calling the `items()` function
- IMPORTANT: `min_heap` will store `(freq, num)`
- if length of `min_heap` is greater than k, pop from it
- RESULT: iterate through `min_heap` and return the 

```python
from collections import Counter
import heapq

nums = [1, 1, 1, 2, 2, 3]
k = 2

count = Counter(nums)

min_heap = []

for num, freq in count.items():
    heapq.heappush(min_heap, (freq, num))
    
    if len(min_heap) > k:
        heapq.heappop(min_heap)

result = [val for freq, val in min_heap]
print(result)
```

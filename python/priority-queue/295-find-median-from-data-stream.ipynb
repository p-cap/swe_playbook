### 295. Find Median from Data Stream

- Data Structure: heap
- The class will have two lisits
- First list will store the lower numbers
- Second list will store the higher numbers
- There will be two functions for the class, addNum and findMedian
- addNum:
    - heapify lower list into max_heap which means the largest number in the lower list can be technically the median if combined with the other list
    - check if the first element of the lower list is higher than the first element of the upper list and if so, we pop the element from the lower list and push onto the upper list
    - we check for unbalance between the upper and lower list
    - who has two or more elements will have an element removed and move to the other list
- findMedian
    - whoever's length is larger than the other, the first element of either list is returned
    - if the lengths are equal, we calculate by adding the first elements of each list and getting the average

```python
import heapq

class MedianFinder:
    def __init__(self): 
        self.lower = []
        self.upper = []
    

    def addNum(self, num):
        heapq.heappush(self.lower, -num)

        if self.lower and self.upper and (-self.lower[0] > self.upper[0]):
            val = -heapq.heappop(self.lower, val)
            heapq.heappush(self.upper, val)
        
        if len(self.lower) > len(self.upper) + 1:
            val = -heapq.heappop(self.lower, val)
            heapq.heappush(self.upper, val)


        if len(self.upper) > len(self.lower) + 1:
            val = -heapq.heappop(self.upper)
            heapq.heappush(self.lower, val)


    def findMedian(self):
        if len(self.lower) > len(self.upper):
            # return first element vs the popped element
            # would that be technically the same?
            return float(-self.lower[0])

        elif len(self.upper) > len(self.lower):
            return float(self.upper[0])
        
        else:
            return (-self.lower[0] + self.upper[0]) / 2
```

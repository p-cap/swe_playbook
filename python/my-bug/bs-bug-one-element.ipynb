### The tale of the one element test case in binary search

We have `nums` with one element and the `target` is within the `nums` element. Using `binary search`, we expect the return of 0

```python
class Solution:
    def bs_search(self, nums:list[int], target:int):
        n = len(nums)
        left, right = 0, n - 1
        while left < right:
            mid = (left + right + 1) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] > target:
                right -= 1
            else:
                left += 1 
        return -1
```

```python
nums = [5]
target = 5
expected = 0
my_solution = Solution()
print(my_solution.bs_search(nums, target))
```

The return is `-1`. This is incorrect because the expected is 0. There is a bug in the code. 
WHen passing nums and target, it never gets into the while loop because left is NOT less than right. 

Replace `while left < right` with `while left <= right`

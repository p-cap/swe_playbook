### Koko Eating Bananas

NOTE: I will try my best to understand the problem
- initialize low and high
    - low = 1 ????
    - high is the maximum element in the list
- set the result to the high
- start the binary search by comparing between low and high
- k, for this problem, is the mid
- total_time is tracked which determines on how our pointer will move based on its values
- iterate through each pile
- track the total time by getting the quotient between the current pile and k and pass to the math.ceil
- compare total_time to h
    - if total_time is lower than h
        - set the k to the result
        - move the high
    - if total_time is higher than h, move the high
- return the high

```python
piles = [3,6,7,11]
h = 8
expected = 4
```

```python
import math

low, high = 1, max(piles)
result = high

while low <= high:
    # k is the mid for the binary search
    k = (low + high + 1) // 2
    
    total_time = 0
    
    for p in piles:
        # math.ceil will always round up even with the slightest decimal
        total_time += math.ceil(p / k)
        
    if total_time <= h:
        result = k
        high = k - 1
    else:
        low = k + 1
    
print(result == expected)
```

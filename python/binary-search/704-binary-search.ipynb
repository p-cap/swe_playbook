### 704 Binary Search

- make sure that the array is sorted
- looking for the mid is add the pointers plus 1 interger divided by 2
- check if the mid index is equal to the target and if so, return the index
- decrement the right pointer if target is greater than mid pointer
- increment the left pointer if the target is less than the left pointer

```python
class Solution:
    def search(self, nums:list[int], target: int):
        n = len(nums)
        left, right = 0, n - 1

        while left <= right:
            mid = (left + right + 1) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left += 1
            else:
                right -= 1
        return -1
```

```python
my_solution = Solution()
nums = [5]
target = 5
print(my_solution.search(nums,target))
```

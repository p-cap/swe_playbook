### Comparing Combination problems
The combination problem is the base problem for Combination Sum and Combination Sum II. The combination problem looks for all elements based on a given k. By defination, the combination will have a set of elements based on from the input array and the input size. A combination does not consider order. For example, [2,3] and [3,2] are the same. The algorithm for finding the combination for a given input array and an input size is backtracking. The backtracking funciton will iterate through the input element, add an element to the list, call the backtrack function passing the incremented index and list. Eventually, the recursive call of the backtrack function will end based on the base case. The base case for combination will check if the list size is equal to the input size and if it is, we add that list to the result list. The result list is the collection of combinations from the input array given a specific size.

In combination, we are passing `i+1` on every call to ensure we do not get any duplicates. 

```python
input_array = [1,2]
k = 1
expected = [[1], [2]]

result = []

def backtrack(start, combination):
    if len(combination) == k:
        result.append(combination[:])
        return
    for i in range(start, len(input_array)):
        combination.append(input_array[i])
        backtrack(i + 1, combination)
        combination.pop()

backtrack(0, [])
print(result)
```

For combination sum, we want to look for combinations that add to a specific sum. The problem states that it is acceptable for repeat characters to be used in the combination. The repeat characters can be found based on the start parameter being passed but not being incremented. Also, if we want to compute for the matched sum, we pass a remaining parameter that is decremented everytime on each call. 

```python
candidates = [2,3,6,7] 
target = 7
output = [[2,2,3],[7]]

# colects all the combinations 
result = []

def backtrack(start,combinations,remaining):
    if remaining == 0:
        result.append(combinations[:])
        return
    elif remaining < 0 or remaining > target:
        return
    for i in range(start, len(candidates)):
        combinations.append(candidates[i])
        # print(remaining - candidates[i])
        backtrack(i, combinations, remaining - candidates[i])
        combinations.pop()
backtrack(0,[],target)
print(result)
```

The combination sum will NOT have an increasing index because it allows the combination elements to be duplicate. 
`backtrack(i, combinations, remaining - candidates[i])`
Unlike, the base combination problem, duplicate characers are not allowed which passes a starting index to increase by 1
`backtrack(i + 1, combination)`

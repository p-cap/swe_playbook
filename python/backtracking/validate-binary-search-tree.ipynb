- call validate function starting the lowest int and the highest int and the root

```python
# root = [2,1,3]
# result = True
def is_valid_bst(root):
    def validate(node, low, high):
        if not node:
            return True

        # I forgot the "not" and that's why I was confused
        if not (low < node.val < high):
            return False
        
        # low will always be float('-inf') and high will always be float('inf')
        return (validate(node.left,low,node.val) and validate(node.left,node.val,high))

return validate(root, float('-inf'), float('inf'))
```

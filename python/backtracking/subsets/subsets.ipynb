### Subset vs Subset II
PROBLEM: Given an array, we need to find all possible powersets of an array. Essentially, we are storing all possible selection of elements based on the given array. 

BASE SOLUTION: The subset problems uses backtracking. The backtrack function takes a start parameter and the path. The start parameter ensures on every recursive call, we have a different start while the path parameter tracks the elements for each subset. The backtrack function does not necessarily have a base case. It collects the path list every single time. The for loop contains the path variable adding elements to it, calling the backtrack function and then popping from the path list. The initial call to the backtrack function is `backtrack(0,[])` which is the starting point of the for loop and the initial path variable

| Feature | Subsets I | Subsets II |
| :--- | :--- | :--- |
| **Input Type** | Unique elements (e.g., `[1, 2, 3]`) | Contains duplicates (e.g., `[1, 2, 2]`) |
| **Pre-processing** | None needed | **Must `nums.sort()` first** |
| **Base Case** | **Collection point:** `res.append(list(path))` is called at the start of every function call. | **Collection point:** `res.append(list(path))` is called at the start of every function call. |
| **Function Signature** | `def backtrack(start, path):` | `def backtrack(start, path):` |
| **Core Loop Logic** | `for i in range(start, len(nums)):`<br>`    path.append(nums[i])`<br>`    backtrack(i + 1, path)`<br>`    path.pop()` | `for i in range(start, len(nums)):`<br>`    if i > start and nums[i] == nums[i-1]: continue`<br>`    path.append(nums[i])`<br>`    backtrack(i + 1, path)`<br>`    path.pop()` |
| **Result Set** | All possible combinations | Only **unique** combinations |

Subset II

```python
nums = [1,2,2]
output = [[],[1],[1,2],[1,2,2],[2],[2,2]]
result = []

def backtrack(start, path):
    result.append(path[:])
    for i in range(start, len(nums)):
        # i > start ensures once i is not the beginning of the loop, it can start comparing
        # this is more of an optimization or 
        if i > start and nums[i-1] == nums[i]:
            continue
        path.append(nums[i])
        backtrack(i + 1, path)
        path.pop()
backtrack(0,[])
print(result)
```

### Permutation v2
- trying to make the problem similar to subset in order to create a pattern
- for permutation, we do NOT pass an different index after each recursive call

```python
nums = [0,1]
result = []
path = []

def backtrack():
    if len(nums) == len(path):
        result.append(path[:])
        return
    
    for i in range(len(nums)):
        if nums[i] in path:
            continue
        path.append(nums[i])
        backtrack()
        path.pop()

backtrack()
print(result)
```

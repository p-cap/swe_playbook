### Why is the `i` in the 17. Letter Combinations of a Phone Number problem is causing a OutofRangeIndexException
The code shows the `i` is causing an index error

```python
class Solution:
    def letterCombinations(self, digits: str) -> list[str]:
        result = []
        number_to_letter = { 
            "2" : "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "pqrs",
            "8": "tuv",
            "9": "wxyz"
        }

        def backtrack(i, combo):
            if len(combo) == len(digits):
                print("here")
                combo.append("".join(combo))
                return
            print(i)
            current_digit = digits[i]
            for letter in number_to_letter[current_digit]:
                print(letter)
                combo.append(letter)
                backtrack(i + 1, combo)
                combo.pop()
        backtrack(0, [])
        return result
    
    solution = Solution()
    digits = "23"
    result = solution.letterCombinations(digits)
    print(result)
```

In line 21, `current_digit = digit[i]` is causing an IndexError. This code is responsible for setting the current digit based on the input. Below is the output of the print statement. We are printing `i` and the letter based on the letter combination mapping. The index would cause an error when `i = 2` because the size of the given digit or string is 2 which does not exists in the `current_digit = digits[i]` to go IndexError.

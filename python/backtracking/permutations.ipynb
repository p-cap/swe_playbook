### Permutations
- initialize:
    - `results` to save all permutations
    - `path` for tracking current permutation
    - `used` with boolean values set to `false`
- backtrack function
    - base case: if path length is equal to length of nums, we add a shallow copy of path to result
    - call for loop
    - check if used is true and if so, go to the next iteration
    - set used current index to true
    - call backtrack
    - pop from path
    - set used index to false after backtrack call

```python
nums = [1,2,3]

result = []
path = []
n = len(nums)
used = [False] * n

def backtrack():
    if len(path) == len(nums):
        result.append(path[:])
        return

    for i in range(n):
        if used[i]:
            continue
        
        used[i] = True
        path.append(nums[i])

        backtrack()

        path.pop()
        used[i] = False
backtrack()
print(result)
```

Note to consider:
- the for loop always start at 0 at every recursive call
- unlike Subset, the for loop allows to re-examine every number for every "slot" in the permutation

### Redo
- initialize result and path
- the used list asks the question, what is left to use?
- BASE CASE: if the path is eqaul to the length of nums
- call the for loop
- if the element is used, move to the next element
- now the element is used, set ith used to true
- append ith number to path
- call the backtrack
- pop from the path
- set the ith index based to false so it can be re-used on the next recursive function

```python
nums = [1,2,3]
path = []
result = []
n = len(nums)
used = [False] * n

def backtrack(): 
    if len(path) == len(nums):
        # again, I FORGOT THE SHALLOW COPY
        result.append(path[:])
        return
    for i in range(len(nums)):
        if used[i]:
            continue

        used[i] = True
        path.append(nums[i])

        backtrack()

        path.pop()
        used[i] = False

backtrack()
print(result)
```

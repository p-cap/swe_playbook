### Permutations
- initialize:
    - `results` to save all permutations
    - `path` for tracking current permutation
- backtrack function
    - base case: if path length is equal to length of nums, we add a shallow copy of path to result
    - call for loop
    - check if the current `nums[i]` is in the `current_path`, move to the next iteration
    - call backtrack
    - pop from path
    - set used index to false after backtrack call

```python
nums = [1,2,3]

result = []
path = []
n = len(nums)

def backtrack():
    if len(path) == len(nums):
        result.append(path[:])
        return

    for i in range(n):
        if nums[i] in path:
            continue
        path.append(nums[i])
        backtrack()
        path.pop()
    
backtrack()
print(result)
```

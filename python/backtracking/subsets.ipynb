### 78. Subsets

```python
nums = [1,2,3]

# this is the final array
result = []
subset = [] # collects each subset

def backtrack(i):
    if i >= len(nums):
        # shallow copy
        result.append(subset[:])
        return

    subset.append(nums[i])
    backtrack(i + 1)

    # preserves the state of subset within the scope of the function call
    subset.pop()
    backtrack(i + 1)

print(result)
```

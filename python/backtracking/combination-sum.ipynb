### 39. Combination sum
- call `backtrack(0,0)`
- if `total` is equal to target, we add `path` to `result`
- if `i` is greater or equal than the length of `path` or if `total` is greater than `target`, we call return
- append ith candidates to `path`
- we call `backtrack` with the current `i` and `i` within candidates plus current total
- we pop from the `path`
- we call `backtrack` with the current `i` plus 1 and the `total`

```python
candidates = [2,3,6,7]
target = 7

result = []
path = []

def backtrack(i, total):
    if total == target:
        result.append(path[:])
    
    if i >= len(candidates) or total > target:
        return
    
    path.append(candidates[i])

    backtrack(i, candidates[i] + total)

    path.pop()

    backtrack(i + 1, total)

backtrack(0,0)
print(result)
```

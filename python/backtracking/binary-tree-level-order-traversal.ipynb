### Binary Tree Level Order Traversal
- `from collections import deque`
- initialize `queue` with iterable root `[root]` as an iterable
- iterate through the `queue`
- initialize `current_level` to append to result and distinguish each level
- iterate through level based on the size of the queue which is the level size
- pop from the `queue`
- append the node value onto the `current_level` 
- check if node left or right is present
- if so, add it to the queue
- outside the for loop, add `current_level` to the `result` array
- return `result`

```python
from collections import deque

def traversal(root):
    if not root: return []

    queue = deque([root])
    result = []
    while queue:
        level_size = len(queue)
        current_level = []

        for _ in range(level_size):
            node = queue.popleft()
            current_level.append(node.val)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        result.append(current_level)
```

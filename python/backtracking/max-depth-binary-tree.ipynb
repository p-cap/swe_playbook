### Max depth binary tree
- base case: if current node is empty, return 0
- return 1 plus the max between the left node function call and right node function call

```python
def max_depth(root):
    if not root:
        return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))
```

```python
from typing import Optional

class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            # having just the return will return a NoneType causing en TypeError
            # return
            return 0
        # root.left is the correct way to call the left attribute of the current node
        # self.maxDepth is the correct way of calling itself
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
```

### Subsets
#### Problem Desription
The problem is looking for all possible subsets given an array. Based on the output of the answer, subset is also called power set. For example, an empty list, [], is considered a subset of a given list. The problem constraints are as follows. The length of the input list has a minimum of 1 and a maximum of 10. Next, the elements within the input has a minimum of -10 and a maximum of 10. Finally, all of the values are unique. The values being unique does help understand that the subset of the input list will not have duplicate values within the same subset. 

### GEMINI'S RECOMMENDED SOLUTION
The recommended solution for subset is to have the backtrack function as an internal function to the parent function of the Solutions class. First, the result list will have all the subsets for the given input list. Next, the declaration of the backtracking function will have an underscore to signify that the function is internal and its a helper function. PEP 8, referred as naming converntions, suggests the underscore for recursive patterns. 

### TIME AND SPACE COMPLEXITY
Time complexity is O(N*2^N). Based on what Gemini said, N is the array size and 2 is the choices of including or excluding. 
We start with an empty set. Below the empty set, which is the next level in the tree, do we include the first element or not? The left branch has the element we included and on the right branch we have the included element and so on.

### IndexError again

```python
nums = [100,4,200,1,3,2]

n = len(nums)

longest = 1

nums.sort()

for i in range(n):
        print(f'i: {i}')
        # IndexError
        if i < n and nums[i] + 1 == nums[i + 1]:
                longest += 1
print(longest)
```

Line 12 is the guilty party. `i<n` and `nums[i+1]` is causing the error. When `i == n - 1`, this still satisfies the condition `i < n` however, `i == n - 1` causes `nums[i+1]` which becomes `nums[n]` causes the `IndexError`

```python
nums = [100,4,200,1,3,2]

n = len(nums)

longest = 1

nums.sort()

for i in range(n):
        # i + 1 should be assessed at the beginning
        if i + 1 < n and nums[i] + 1 == nums[i + 1]:
                longest += 1
print(longest)
```

`i + 1 < n` ensures that when i is added with 1, it will not be equal or more than `n`

#### Classes and init

```python
class Codec:
    def __init__(self):
        delimiter = '\n'
```

```python
test = Codec()
print(test.delimiter)
```

In your code, delimiter is not an attribute because it is defined as a local variable rather than being "attached" to the instance.

To turn a variable into an attribute in Python, you must prefix it with self..

```python
class Codec:
    def __init__(self):
        self.delimiter = 'wassup'

test = Codec()
print(test.delimiter)
```

### The difference between defaultdict and standard Python library

A defaultdict is a specialized version of the standard Python dictionary found in the collections module. Its "superpower" is that it provides a default value for a key that does not yet exist, rather than crashing your program with a KeyError.

#### Testing defaultdict library

```python
from collections import defaultdict

my_defaultdict = defaultdict(list)
```

my_defaultdict['nokey'] -> this key does not exists

```python
my_defaultdict['nokey']
```

printing the actual defaultdict 


```python
print(my_defaultdict)
```

#### Testing standard python dictionary

```python
mydict = {}
```

accessing key that does not exists in a standard defaultdict

```python
mydict['nokey']
```

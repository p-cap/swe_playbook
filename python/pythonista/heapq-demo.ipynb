#### Using heapq library

```python
from collections import Counter
import heapq

nums = [1,2,1,2,1,2,3,1,3,2]
minHeap = []
k = 2
```

- cast the nums array into a Counter

```python
mycounter = Counter(nums)
print(mycounter)
```

- iterate through the array and start the heapify process

```python
for key,value in mycounter.items():
    heapq.heappush(minHeap,(value, key))
    
    if len(minHeap) > k:
        heapq.heappop(minHeap)

print(minHeap)    
```

- Taking the key from the minHeap array using list comprehension

```python
ret = [key for val, key in minHeap]
print(ret)
```

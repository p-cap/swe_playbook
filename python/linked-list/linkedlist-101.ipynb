- `append` function

    - check for `self.head` and if it is set to `None`, it will point it to the new node
    - set `current` to `self.head`
    - iterate through the linkedlist while `current.next` is not `None`

```python
# Color Constants
GREEN = '\033[92m'
YELLOW = '\033[93m'
RED = '\033[91m'
BLUE = '\033[94m'
RESET = '\033[0m'

class Node:
    # def __init__(self)
    # Above was the issue because it would not allow for Node class to take another argument based on new_node = Node(value)
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        print(f"\n{RED}[Action]{RESET}: Appending {value}")
        new_node = Node(value)
        
        if not self.head:
            print(f"  -> List was empty. Setting head to Node({value})")
            self.head = new_node
            return

        current = self.head
        print(f"  -> Starting at head: Node({current.value})")
        
        # This while loop only triggers from the 3rd element onwards
        step = 1
        while current.next:
            print(f"  -> Step {step}: Node({current.value})'s next is not None (it's Node({current.next.value})). Moving forward...")
            current = current.next
            step += 1
        
        print(f"  -> Found the end! Current is Node({current.value})")
        print(f"  -> Linking: Node({current.value}).next = Node({value})")
        current.next = new_node
        
    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(str(current.value))
            current = current.next
        print("-> ".join(elements) + "-> None")

# instantiate the linkedList            
ll = LinkedList()

# append the values
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

# display the elements
ll.display()
```

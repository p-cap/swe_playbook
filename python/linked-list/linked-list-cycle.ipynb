### 141. Linked List Cycle
- IMPORTANT: initialize a hash set 
- iterate through the linked list 
- if `node` in hashset, return `true`
- otherwise, add `node` to hashset
if while loop ends, return `false`

```python
class Node:
    def __init__(self,val):
        self.value = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self,val):
        new_node = Node(val)

        if not self.head:
            self.head = new_node
            return
        
        current = self.head

        # get a better understanding if using current or current.next
        while current.next:
            current = current.next

        current.next = new_node

    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(current.value)
            current = current.next
        print(elements)

    def make_cycle(self,val=1):
        current = self.head
        next_node = None
        hash_set = set()
        while current.next:
            if current.value == val:
                next_node = current
            if current.value in hash_set:
                break
            else:
                hash_set.add(current.value)
            current = current.next
        current.next = next_node
        check_node = self.head
        hash_set = set()
        terminate = False
        while not terminate:
            print(f'Node({check_node.value}) -> Node({check_node.next.value})')
            if check_node.value in hash_set:
                print('inside the if')
                terminate = True
            else:
                hash_set.add(current.value)
            check_node = check_node.next

        

ll = LinkedList()
ll.append(1)
ll.append(3)
ll.append(5)
ll.display()
ll.make_cycle()
ll.display()
```

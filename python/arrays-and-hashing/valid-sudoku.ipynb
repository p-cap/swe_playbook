### 36. Valid Sudoku

BREAKDOWN:
- declare 3 defaultdict
- take the val from the board
- if the value is '.', continue to the iteration
- check if the value is in any of the 3 defaultdict
- add row, col and squares value

```python
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
```

- initialize defaultdict that has set value

```python
from collections import defaultdict

rows = defaultdict(set)
cols = defaultdict(set)
squares = defaultdict(set)
```

- iterate through the array
- check the value in rows, cols or squares
- rows[r].add(value) -> add row value
- cols[c].add(value) -> add cols value
- square[(r // 3, c //3)].add(value) -> (r // 3, c // 3) -> square elements

```python
for r in range(len(board)):
    for c in range(len(board[0])):
        value = board[r][c]
        
        if value == '.':
            continue
        
        if value in rows[r] or value in cols[c] or value in squares[r // 3, c // 3]:
            print(False)
        
        rows[r].add(value)
        cols[c].add(value)
        squares[(r // 3, c // 3)].add(value)
        
print(True)    
```

```python
test = [3,0,1,0]

from collections import defaultdict

temp_array = []
final = []
mydict = defaultdict(int)
k = 1
```

- Built the defaultdict to tally the elements

```python
for num in test:
    mydict[num] += 1
print(mydict)
```

- Iterate through mydict
- appended val and key in that order
- val and key in this order allows us to sort it
- call sort(reverse=True)

```python
for key, val in mydict.items():
    temp_array.append((val,key))
print(f'Before: {temp_array}')
temp_array.sort(reverse=True)
print(f'After: {temp_array}')
```

- add the key to the final array
- return the slice of the final array based on k

```python
for val,key in temp_array:
    final.append(key)
print(final[:k])
```

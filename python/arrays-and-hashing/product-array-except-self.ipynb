### 238. Product of Array Except Self

BREAKDOWN
- iterate nums from left to right
- multiply left to the output element
- set the left with the profuct between the left and the current element
- iterate nums from right to left
- multiply right variable to the current element of the output 
- update the right with the current element

```python
test = [1,2,3,4]
output = [1] * len(test)
```

- iterate nums from left to right
- multiply left to the output element
- set the left with the profuct between the left and the current element

```python
left = 1
for i in range(len(test)):
    output[i] *= left
    left *= test[i]
```

- iterate nums from right to left
- multiply right variable to the current element of the output 
- update the right with the current element

```python
right = 1
for i in range(len(test) - 1, -1, -1):
    output[i] *= right
    right *= test[i]
```

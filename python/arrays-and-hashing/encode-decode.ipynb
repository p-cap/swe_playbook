### 271. Encode and Decode Strings

Breakdown
- encode
    - determine the letter separator
    - determine the end for each word
    - ord(ch) convert to string and add it the overall string
- decode
    - look for the separator
    - cast the string to int and str again 

```python
from typing import List

class Codec:
    def encode(self, strs: List[str]) -> str:
        encoded_string = ''
        letter_sep = ':'
        end = '#'

        for word in strs:
            for ch in word:
                encoded_string += str(ord(ch)) + letter_sep
            encoded_string += end
        return encoded_string

    def decode(self, s: str) -> List[str]:                
        letter = ''
        decoded_string = ''
        end = '#'
        result = []
        for char in s:
            if char == ':':
                decoded_string += chr(int(letter))
                letter = ''
            elif char.isdigit():
                letter += char
            elif char == end:
                result.append(decoded_string)
                decoded_string = ''
        return result
```

```python
strs = ['Hello', 'World']

encoder = Codec()
print(f'strs: {strs}')
msg = encoder.encode(strs)
print('encoding msg')
print(f'msg: {msg}')
print('decoding msg')
decoded_msg = encoder.decode(msg)
print(f'decoded_msg: {decoded_msg}')
```

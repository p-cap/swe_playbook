### Remove Duplicate Letter using Monotonic stack
- initialize:
    - stack -> this will be returned and stores the char preserving lexicographical order and its order within the original list
    - visited -> this tracks chars that were seen and can store unqualified chars at the time of iteration so it can be revisited again
    - last_occ (hashmap) -> tracks the last occurence of the chars 
- iterate through the string accessing the index and char
- check if the char is in visited
- the while loop checks
    - if stack is not empty
    - if the current char is less than the top of the stack
    - if the current index is less than the index of the stack
- append current char to the stack 
- add char to visited
- `index < last_occ[stack[-1]]` 
    - this follows up with char < stack[-1]
    - the index of that char needs to be less than the index of the top of the stack 

```python
s = "bcabc"

last_occ = { c : i for i, c in enumerate(s) }

stack = []
visited = set()

for index, char in enumerate(s):
    if char in visited:
        continue
    
    while stack and char < stack[-1] and index < last_occ[stack[-1]]:
        popped_char = stack.pop()
        visited.remove(popped_char)
    
    stack.append(char)
    visited.add(char)

print(stack)
    
```

```python
for i in range(len(nums)):
    if i > 0 and nums[i] == nums[i-1]:
        continue
    
    j = i + 1
    k = len(nums) - 1
    
    while j < k:
        total = nums[i] + nums[j] + nums[k]
        if total > 0:
            k -= 1
        elif total < 0:
            j += 1
        else:
            res.append((nums[i], nums[j], nums[k]))
            while nums[j] == nums[j- 1] and j < k:
                j += 1

print(res)
```

Recall:
- sort the list
- iterate through the list
- check if the val is greater than zero and if so, terminatee the for loop
- check for duplicates by ensuring the index is greater than 0 and if the current element is equal to the previous element and if so, move to the next iteration
- call the two_sum function which receives the following parameters
    - list input
    - list storing the triplets
    - the current index within the for loop
- In the two_sum function, we initialize our pointers
- NOTE: left pointer will be one spot ahead of the input index
- iterate through the pointers by checking the sum of the 3 elements
- if the sum is greater than 0, move the right pointer
- if the sum is less than 0, move the left pointer
- else, we can append the three elements into the result list
- move both pointers
- check for duplicate elements by checking if the current left is the same as the previous left and of course, it falls within the left < right constraint
- move the left pointer until it breaks the while loop 
- return the result list containg the triplets

```python
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        result = []
        nums.sort()
    
        for index, val in enumerate(nums):

            # terminates the loop if the val is greater than 0
            if val > 0:
                break

            # skip neighboring duplicates
            # this will organically happen because of the sorting
            if index > 0 and val == nums[index - 1]:
                continue
            
            self.two_sum(nums,index,result)
            
        return result

    def two_sum(self, nums, first, result):
        n = len(nums)
        # this is sad, I did not catch this
        # left is suppose to be different fromt first
        left = first + 1
        right = n - 1

        while left < right:
            my_sum = nums[first] + nums[left] + nums[right]
            if my_sum < 0:
                left += 1
            elif my_sum > 0:
                right -= 1
            else:
                result.append([nums[first], nums[left], nums[right]])
                left += 1
                right -= 1
                while  nums[left] == nums[left - 1]  and left < right:
                    left += 1
                
        return result 
```

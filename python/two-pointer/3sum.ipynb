### 3sum
- ALGO: Two Pointer
- initialize the `result` to save the collection of lists that sum up to zero
- sort the input list
- iterate the the input list
- if the element is greater than zero, break the for loop 
```python
if nums[0] > 0:
    break
```
- if the index is greater than zero and if the current nums and the previous nums is the same, go to the next iteration
```python
if i > 0 and nums[i] == nums[i-1]:
    continue
```
- create and call the `two_sum` function that accepts the anchor index
- initialize the following:
```python
left = i + 1 # left pointer is beside the anchor index
roght = len(nums) 
```
- use the while loop with `left < right` expression
- calculate the `my_sum = i + left + right`
- if `my_sum` is greater than 0, move the `right` pointer
- if `my_sum` is less than 0, move the `left` pointer
- else, add the 3 elements to the result list
- move the both left and right pointer
- again, check if the current left and previous left pointer, move the left pointer

````python
class Solution:
    def threeSum(self, nums: list[int]) -> list[list[int]]:
        result = []
        n = len(nums)
        
        def two_sum(first):
            second = first + 1
            third = len(nums) - 1

            while second < third:
                my_sum = nums[first] + nums[second] + nums[third]
                if my_sum < 0:
                    second += 1
                elif my_sum > 0:
                    third -= 1
                else:
                    result.append([nums[first], nums[second], nums[third]])
                    second += 1
                    third -= 1
                    while second < third and nums[second] == nums[second - 1]:
                        second += 1
        nums.sort()
        print(nums)
        for i in range(n):
            # I do not need i > 0 for nums[i]
            # if i > 0 and nums[i] > 0:
            if nums[i] > 0:
                break
            elif i > 0 and nums[i-1] == nums[i]:
                continue
            else:
                two_sum(i)
        return result 
````

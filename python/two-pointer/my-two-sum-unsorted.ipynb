### My Two Sum with unsorted input 
- initialized a list 
- the list will store the value and the index within a tuple
- having the value as the first element is crucial in order to sort the values but keep its corresponding index
- `sort()` the list of tuples. By default it will sort the first element
- initialize the two pointers
- check the sum between the first elements of the tuple
- move the pointer accordingly based on the comparision between target and sum

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        n = len(nums)
        left = 0
        right = n - 1
        my_list = []

        for i, val in enumerate(nums):
            my_list.append((val, i))
        
        my_list.sort()
        
        while left < right:
            my_sum = my_list[left][0] + my_list[right][0]
            if  my_sum == target:
                index_left = my_list[left][1]
                index_right = my_list[right][1]
                return [index_left, index_right]
            elif my_sum > target:
                right -= 1
            else:
                left += 1 
                
my_solution = Solution()
nums = [2,7,11,15]
target = 9
result = my_solution.twoSum(nums,target)
print(result)
```

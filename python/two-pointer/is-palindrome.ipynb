IsPalindrome 
- initialize the left and right pointer
- use a while loop with the expressions:
    - `left <=right`
    - `left < right`
- if the current char is not a char move left or right pointer
- also, move to the next iteration with the `continue` keyword
- if the left and right pointer are NOT equal, we can return False
- otherwise move both pointers

```python
s = "A man, a plan, a canal: Panama"
left = 0
right = len(s)
result = True

while left <= right:
    if not s[left].isalnum():
        left += 1
        # I forgot to move to the next iteration
        continue
    
    if not s[right].isalnum():
        right -= 1
        continue

    if s[left] != s[right]:
        result = False
        break
    else:
        left += 1
        right -= 1

print(result)
```

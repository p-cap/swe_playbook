```python
height = [1,8,6,2,5,4,8,3,7]
```

```python
left = 0
right = len(height) - 1
max_v = 0

while left < right:
    width = right - left
    h = min(height[left], height[right])
    current_area = h * width
    max_v = max(max_v, current_area)
    
    if height[left] < height[right]:
        left += 1
    else:
        right -= 1

print(max_v)
```

```python
height = [1,8,6,2,5,4,8,3,7]
left, right = 0, len(height) - 1
max_area = 0
```

```python
while left < right:
    min_height = min(height[left], height[right])
    max_area = max(max_area, (right - left) * min_height)
    if height[left] < height[right]:
        left += 1
    else:
        right -= 1
print(max_area)
```

### Got it right again
- intiailized two pointers
- iterate `left < right`
- calculate the current distance based on right minus left
- get the min height between the left and right heights
- get the max area based on the distance multiplied by the min_height
- Why move the pointer? 
    - Because we are looking for the max possible value, we can compare left and right heights
    - whoever ever is lesser will move because we need max area
    - lesser values need to find bigger values by moving

### Repitiion is key

```python
class Solution:
    def maxArea(self, height: list[int]) -> int:
        # two pointer opposite direction 
        # calculate the distance between the two points
        # get the minimum between the two points 
        # the max area is the product between minimum of the two points and the distance
        # which ever pointer that has a lesser value moves
        n = len(height)
        left = 0
        right = n - 1
        max_area = 0
        while left < right:
            distance = right - left
            minimum_height = min(height[left], height[right])
            max_area = max(max_area, minimum_height * distance)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
```

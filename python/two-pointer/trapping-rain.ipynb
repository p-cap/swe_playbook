### Trapping Rain Water

```python
height = [0,1,0,2,1,0,1,3,2,1,2,1]
```

- initialize with the leftmost and rightmost elements
- initialize both elements as left and right maxes
- iterate through using the pointers
- if left_max is less than right_max, we move the left pointer, otherwise, the right pointer
- we increment or decrement the pointer based on the left and right values being lesser
- take the max between the current max and the next height
- add the difference the max value and the current height

```python
if not height: print(0)

n = len(height)
left, right = 0, n - 1
left_max = height[left]
right_max = height[right]
result = 0

while left < right:
    # why is this the condition
    # the height matters because we need the minimum height
    if left_max < right_max:
        left += 1 
        left_max = max(left_max, height[left])
        result += left_max - height[left]
    else:
        
        right -= 1
        right_max = max(right_max, height[right])
        result += right_max - height[right]

print(result)
```

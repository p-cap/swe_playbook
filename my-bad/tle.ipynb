### Time Limited Exceeded

```python
nums = [100,4,200,1,3,2]
```

```python
# WHY TLE
n = len(nums)
longest = 0

for i in range(n):
# nums[0] - 1 = 99
  if nums[i] - 1 not in nums:
        longest = 1
        # nums[0] + 1 = 100
        while nums[i] + 1 in nums:
            # 100
            # nums[i] + 1 is always 100
            longest += 1
print(longest)
```

In line 10, `nums[i] + 1` will never change. It'll be always `nums[i] + 1`

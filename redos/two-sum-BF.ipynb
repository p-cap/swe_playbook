### Two sum - REDO (BF)

```python
nums = [2,5,5,11]
target = 10
```

- the outer for loop interates through the nums array
- the inner for loop is plus one to the index of the outer loop

```python
size = len(nums)
for i in range(size):
    for j in range(i + 1, size):
        if nums[i] + nums[j] == target:
            print([i,j])
```

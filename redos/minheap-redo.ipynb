### Heapq demo redo
- given nums = [1,2,1,2,1,2,3,1,3,2]
- we import Counter and heapq
- initialize an empty array as minHeap
- for top k element problem, we would have a k
- instantiate the Counter by passing the nums array
- we iterate through the counter pulling the key and value
- we push the key value as a tuple into the minHeap empty array
- if the length of minHeap is greater than the input k, we heappop from minHeap
- We can return the key from the minHeap using list comprehension
    - we say that we want the value from the decontructing key and value from the minHeap array of tuples

```python
nums = [1,2,1,2,1,2,3,1,3,2]
k = 2
```

- import Counter and heapq

```python
from collections import Counter
import heapq
```

- initialize minHeap
- cast nums array into counter

```python
minHeap = []
mycounter = Counter(nums)
```

- deconstruct key and value from the counter.items()
- heappush tuple into minHeap array
- if the length is greater than two, pop from the minHeap

```python
for key, value in mycounter.items():
    heapq.heappush(minHeap,(value, key))
    if len(minHeap) > 2:
        heapq.heappop(minHeap)
print(minHeap)
```

- take key from the deconstructed minHeap

```python
ret = [key for value, key in minHeap]
print(ret)
```

Encode and Decode String - REDO

- added the __init__ function
- added the self.delimiter attribute
- NOTE: WITHOUT THE SELF, delimiter is considered a local variable to the function
- my encode function works like this:
    - joins the array with a custom delimiter
    - reverses the string
    - pass each character to the ord() function
    - concatenate each character with ':'
    - ':' serves as the delimiter for each character
- my decode function works like this:
    - check if the delimiter is the char and if it is, it is concatenate with the rest of the string and calls continue for the next iteration
    - if char is colon, we concatenate the char to the overall string
    - if both conditions are not met, we just concatenate the string but before that:
        - we cast the char into an int
        - that int is passed into the chr function
        - then we concatenate it to the overall string
    - we reverse the string
    - concatenate it to the overall output variable and this is the decoded value

```python
class Codec:    
    def __init__(self):
        self.array_delimiter = '\n'
    
    def encode(self, strs: List[str]) -> str:
        encoded_string = ''
        temp = self.array_delimiter.join(strs)
        for ch in reversed(temp):
            encoded_string += str(ord(ch)) + ':'
        return encoded_string

    def decode(self, s: str) -> List[str]:                
        mychar = ''
        decoded_string = ''
        output = ''
        for ch in s:
            if ch == self.array_delimiter:
                decoded_string += ch
                continue
            elif ch != ':':
                mychar += ch
            else:
                decoded_string += chr(int(mychar))
                mychar = ''
        
        for ch in reversed(decoded_string):
            output += ch
        return output.split(self.array_delimiter)
```

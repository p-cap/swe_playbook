### Longest Consecutive Sequence - REDO

##### My Solution

```python
nums = [9,1,4,7,3,-1,0,5,8,-1,6]
```

- set nums

```python
nums_set = set(nums)
```

- BIGGEST TAKEAWAY: It does matter when iterating through a set vs the actual nums
- For the if statement, it checks if it does not have a previous number and if it does, it sets the starting length count to 1
- the while loop looks for the incremented number based on length value
- remember, we check if the num - 1 IS NOT IN nums_set. Only 2 numbers satisfy the condition:
    - 3
    - -1

```python
longest = 0
print(f'nums_set: {nums_set}')
for num in nums_set:
    if (num - 1) not in nums_set:
        print(num)
        length = 1
        while num + length in nums_set:
            length += 1
        longest = max(longest, length)
```

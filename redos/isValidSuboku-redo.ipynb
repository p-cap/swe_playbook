```python
board = [[".",".","4",".",".",".","6","3","."],[".",".",".",".",".",".",".",".","."],["5",".",".",".",".",".",".","9","."],[".",".",".","5","6",".",".",".","."],["4",".","3",".",".",".",".",".","1"],[".",".",".","7",".",".",".",".","."],[".",".",".","5",".",".",".",".","."],[".",".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".",".","."]]
```

- create a dictionary of sets that track rows, cols and square

```python
from collections import defaultdict

seen_row = defaultdict(set)
seen_col = defaultdict(set)
seen_square = defaultdict(set)
```

```python
rows = len(board)
cols = len(board[0])

for r in range(rows):
    for c in range(cols):
        val = board[r][c]
        if val == '.': continue
        
        if val in seen_row[r] or val in seen_col[c] or val in seen_square[(r // 3, c // 3)]:
            print(False)
            break
        
        seen_row[r].add(val)
        seen_col[r].add(val)
        seen_square[(r // 3, c // 3)].add(val)

print(True)
```

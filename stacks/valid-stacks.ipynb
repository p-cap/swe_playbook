#### Valid Parenthesis

- initialize stack
- declare a map of open brackets mapped to closed brackets
- iterate through the string
- check if the char in the string is in the brackets map
- if it is, pop an element from the stack
- compare if the current element used as key to the brackets map is equal to the popped element
- if not return False
- if the chat is NOT in the brackets map, add the element to the stack

```python
s = "([)]"
```

```python
brackets = {
    ']':'[',
    '}':'{',
    ')':'('
}
stack = []
```

```python
for ch in s:
    if stack and ch in brackets:
        current = stack.pop()
        if brackets[ch] != current:
            print(False)
            break
    else:    
        stack.append(ch)
```

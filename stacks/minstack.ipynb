#### MInStack

- create a Minstack class
- have a push function that appends to s stack
- have a pop function that pops from the stack
- have a top function that checks the top of the stack
- have a getMin function that checks the min value within the stack

```python
class MinStack:

    def __init__(self):
        self.stack = []        

    def push(self, val: int) -> None:
        self.stack.append(val)

    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[len(self.stack) - 1]
    def getMin(self) -> int:
        mini = float('inf')
        for element in self.stack:
            mini = min(mini, element)
        return mini 


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(minStack.getMin())
minStack.pop()
print(minStack.top())
print(minStack.getMin())
```

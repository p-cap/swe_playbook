#### Car Fleet

```python
target = 12
position = [10,8,0,5,3]
speed = [2,4,1,1,3]
```

- combine the position and speed
- the position is the first element and speed is the second element
- we sorted the combination in a descending order
- the position will be the reference of the sort
- declare a stack to track the calculated time

```python
cars = sorted(zip(position,speed), reverse=True)
stack = []
```

- CRUCIAL: time is calculated based on the difference between the target and the current position of the car divided by the speed 
- if the time is GREATER THAN THE TIME in the top of the stack, add it to the stack
- times that are different indicate that the cars are in a different fleet

```python
for p, s in cars:
    # calculate the time it takes to get to the target based on position and speed
    time = (target - p) / s
    if not stack or time > stack[-1]:
        stack.append(time)
    
print(f'stack: {stack}')
print(f'length of stack: {len(stack)}')
```

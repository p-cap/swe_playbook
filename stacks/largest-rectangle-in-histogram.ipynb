#### Largest Rectangle in Histogram

- initialize a stack
- iterate through the heights array but adding -1 to the array
- -1 ensures that it handles a test case like [1,2,3,4] and pops the elements from stack
- inside the stack is a while look that checks if the stack is not empty and the top of the stack is greater or equal to the current height
- if the top of the stack is greater than or equal to the current height
- we pop the width and the height
- we add the current width and width from the stack
- we take the max between the current max and the product between the current width and current height

```python
heights = [2,1,5,6,2,3]
```

- initialize the stack and result

```python
stack = []
result = 0
```

- add the -1 element to the array
- set the current_width to 0 ???
- check if the top of the stack is greater than the current height
- if so, we pop ths stack
- add the currnet_width and the width 
- take the max between the current_result and the product betwen the current_width and current_height
- stack tracks current width and height 

```python
for h in heights + [-1]:
    current_width = 0
    while stack and stack[-1][1] >= h:
        width, current_height = stack.pop()
        current_width += width
        result = max(result, current_width * current_height)
    
    stack.append((current_width + 1, h))

print(result)
```

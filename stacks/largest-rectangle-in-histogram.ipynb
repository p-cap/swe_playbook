#### Largest Rectangle in Histogram

- initialize a stack
- iterate through the heights array but adding -1 to the array
- -1 ensures that it handles a test case like [1,2,3,4] and pops the elements from stack
- inside the stack is a while look that checks if the stack is not empty and the top of the stack is greater or equal to the current height
- if the top of the stack is greater than or equal to the current height
- we pop the width and the height
- we add the current width and width from the stack
- we take the max between the current max and the product between the current width and current height

### IndexError list index out of range

Problem: I was trying to iterate through an array in reverse, however, I encountered an IndexError

```python
test = [1,1,0,0,2]
n = len(test)
output = 0
```

```python
for i in range(n - 1, -1, -1):
    output += test[i] + test[i + 1]    
```

Given n = 5, accessing test[i+1] which is test[5], is causing the IndexError

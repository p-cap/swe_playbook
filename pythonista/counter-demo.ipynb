### Counter demo

```python
from collections import Counter
```

#### Test case with 0s as the most number of elements

```python
test = [3,0,1,0]
```

```python
mycounter = Counter(test)
print(mycounter)
```

```python
for key, value in mycounter.items():
    print(key)
    
```

### Test case with no 0 element

```python
nums = [1,2,1,2,1,2,3,1,3,2]
```

```python
mynums = Counter(nums)
print(mynums)
```

```python
for key,val in mynums.items():
    print(key)
```

### Using most_common

```python
for key,value in mycounter.most_common():
    print(key)
```
